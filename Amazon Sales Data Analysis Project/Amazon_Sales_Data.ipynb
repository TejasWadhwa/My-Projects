# Now lets convert Order Date and Ship Date in the Datetime format

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
df.dtypes
```

# Now lets slit Order Date into 2 parts as Order Month, Order Year as it helps data for a particular Year and Monthwise

```python
df['Order Month'] = df['Order Date'].dt.strftime('%B')
df['Order Year'] = df['Order Date'].dt.year
```

# Total Profit per Year

```python
yearly_profit = df.groupby('Order Year')['Total Profit'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(data = yearly_profit, x = 'Order Year', y = 'Total Profit', marker = 'o',  markerfacecolor='green',  markersize=10)
plt.title("Total Profit per Year")
plt.xlabel("Year")
plt.ylabel("Total Profit")
plt.grid()
plt.show()
```

# Top 5 Item Types with the Highest Profit

```python
def thousands_formatter(x, pos):
    if x >= 1e6:
        return f'{x*1e-6:.1f}M'
    else:
        return f'{x*1e-3:.1f}K'
        
item_profit = df.groupby('Item Type')['Total Profit'].sum().reset_index()
top_5_items = item_profit.sort_values(by = 'Total Profit', ascending = False).head(5)
plt.figure(figsize=(10,5))
sns.barplot(x = 'Item Type', y = 'Total Profit', data = top_5_items, palette='gist_earth')
plt.title("Top 5 Items with the highest Total Profit")
plt.xlabel("Item")
plt.ylabel("Total Profit")
plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))
plt.show()
```

# Which Sales Channel was mostly used?

```python
sales_channel = df['Sales Channel'].value_counts()
plt.figure(figsize=(12,6))
plt.pie(
    sales_channel,
    labels = sales_channel.index,
    autopct='%1.1f%%',
    colors = sns.color_palette('Set1', len(sales_channel)),
)
plt.title('Mostly Used Sales Channel')
plt.show()
```

# How much Units were sold in the year 2012 per Month

```python
data_2012 = df[df['Order Year'] == 2012]
data_2012['Order Month'] = data_2012['Order Date'].dt.month
units_sold_2012_monthly = data_2012.groupby('Order Month')['Units Sold'].sum().reset_index()
units_sold_2012_monthly = units_sold_2012_monthly.sort_values(by='Order Month')

plt.figure(figsize=(10, 6))
sns.barplot(x='Order Month', y='Units Sold', data=units_sold_2012_monthly, palette='Set1')
plt.title('Units Sold in Year 2012 Month-wise')
plt.xlabel('Month')
plt.ylabel('Units Sold')

month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct',]
plt.xticks(ticks=range(10), labels=month_names)
plt.show()
```

# Visualize the Item Types with their Order Priorities and Total Revenue

```python
def thousands_formatter(x, pos):
    if x >= 1e6:
        return f'{x*1e-6:.1f}M'
    else:
        return f'{x*1e-3:.1f}K'
items = df.groupby('Item Type')['Total Revenue'].sum().index
all_items = df[df['Item Type'].isin(items)]
plt.figure(figsize=(15, 8))
sns.barplot(x = 'Item Type', y = 'Total Revenue', hue = 'Order Priority', data = all_items, palette='cubehelix', errorbar=None, estimator = sum, order = items)
plt.title('Item Types with their Order Priorities and Total Revenue')
plt.xlabel('Items')
plt.ylabel('Total Revenue')
plt.legend(title = 'Order Priority')
plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))
plt.show()
```

# Code in Modular Format

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def my_data(file_path):
    df = pd.read_csv(file_path)
    return df
```

```python
def order_and_ship(df):
    df['Order Date'] = pd.to_datetime(df['Order Date'])
    df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
def month_and_year(df):
    df['Order Month'] = df['Order Date'].dt.strftime('%B')
    df['Order Year'] = df['Order Date'].dt.year
```

```python
def profit_per_year(df):
    yearly_profit = df.groupby('Order Year')['Total Profit'].mean().reset_index()
    plt.figure(figsize=(10,5))
    sns.lineplot(data = yearly_profit, x = 'Order Year', y = 'Total Profit', marker = 'o',  markerfacecolor='green',  markersize=10)
    plt.title("Total Profit per Year")
    plt.xlabel("Year")
    plt.ylabel("Total Profit")
    plt.grid()
    plt.show()
```

```python
def thousands_formatter(x, pos):
        if x >= 1e6:
            return f'{x*1e-6:.1f}M'
        else:
            return f'{x*1e-3:.1f}K'
def top5_items_highest_profit(df):
    
    item_profit = df.groupby('Item Type')['Total Profit'].sum().reset_index()
    top_5_items = item_profit.sort_values(by = 'Total Profit', ascending = False).head(5)
    plt.figure(figsize=(10,5))
    sns.barplot(x = 'Item Type', y = 'Total Profit', data = top_5_items, palette='gist_earth')
    plt.title("Top 5 Items with the highest Total Profit")
    plt.xlabel("Item")
    plt.ylabel("Total Profit")
    plt.gca().yaxis.set_major_formatter(FuncFormatter(thousands_formatter))
    plt.show()
```

```python
def sales_channel_used(df):
    sales_channel = df['Sales Channel'].value_counts()
    plt.figure(figsize=(12,6))
    plt.pie(
        sales_channel,
        labels = sales_channel.index,
        autopct='%1.1f%%',
        colors = sns.color_palette('Set1', len(sales_channel)),
    )
    plt.title('Mostly Used Sales Channel')
    plt.show()
```

```python
def units_sold_in_2012(df):
    data_2012 = df[df['Order Year'] == 2012]
    data_2012['Order Month'] = data_2012['Order Date'].dt.month
    units_sold_2012_monthly = data_2012.groupby('Order Month')['Units Sold'].sum().reset_index()
    units_sold_2012_monthly = units_sold_2012_monthly.sort_values(by='Order Month')
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Order Month', y='Units Sold', data=units_sold_2012_monthly, palette='Set1')
    plt.title('Units Sold in Year 2012 Month-wise')
    plt.xlabel('Month')
    plt.ylabel('Units Sold')
    
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct',]
    plt.xticks(ticks=range(10), labels=month_names)
    plt.show()
```

```python
def unit_price_and_cost(df):
    fig, (f1, f2) = plt.subplots(nrows = 1, ncols = 2, figsize=(12, 6))
    sns.boxplot(x = df['Unit Price'], ax = f1, color = 'pink', width=0.6, )
    f1.set_title('Unit Price Boxplot')
    f1.set_xlabel('Unit Price')
    
    sns.boxplot(x = df['Unit Cost'], ax = f2, color = 'lightgreen', width=0.6)
    f2.set_title('Unit Cost Boxplot')
    f2.set_xlabel('Unit Cost')
    
    plt.tight_layout()
    plt.show()
```
