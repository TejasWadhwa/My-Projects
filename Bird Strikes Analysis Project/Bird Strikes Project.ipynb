# ● Visuals Depicting the Number of Bird Strikes


```python
data['year'] = pd.to_datetime(data['FlightDate']).dt.year
yearly_strikes = data['year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
sns.lineplot(
    x = yearly_strikes.index,
    y = yearly_strikes.values,
    marker = 'o',
    markersize=10,
    markerfacecolor='red'
)
plt.title('Yearly Bird Strikes')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.show()
```

# Yearly Analysis & Bird Strikes in the US


```python
data['FlightDate'] = pd.to_datetime(data['FlightDate'])

data['Year'] = data['FlightDate'].dt.year

non_us_states = ['Quebec', 'British Columbia', 'Prince Edward Island', 'Ontario', 
                  'Newfoundland and Labrador', 'Alberta', 'Saskatchewan']

data_us = data[~data['Origin State'].isin(non_us_states)]  


plt.figure(figsize=(16, 8))
sns.countplot(x='Year', hue='Origin State', data=data_us, palette='Paired')
plt.title('Yearly Analysis & Bird Strikes in the US')
plt.xlabel('Year')
plt.ylabel('Number of Bird Strikes')
plt.legend(title='Origin State', bbox_to_anchor=(1.03, 1.4))
plt.show()
```

# Top 10 US Airlines in terms of having encountered bird strikes

```python
top_10_airlines = data["Aircraft: Airline/Operator"].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x = top_10_airlines.index, y = top_10_airlines.values, palette='Blues_d')
plt.title("Top 10 US Airlines in terms of having encountered bird strikes")
plt.xlabel("Airlines")
plt.ylabel("Number of Bird Strikes")
plt.xticks(rotation=90)
plt.show()
```

# Airports with most incidents of bird strikes – Top 50


```python
top_50 = data["Airport: Name"].value_counts().nlargest(50)
plt.figure(figsize=(14,17))
sns.barplot(x = top_50.values, y = top_50.index, palette = "CMRmap")
plt.title("Airports with most incidents of bird strikes – Top 50")
plt.xlabel("Number of Bird Strikes")
plt.ylabel("Airports")
plt.show()
```

# Yearly Cost Incurred due to Bird Strikes:

```python
yearly_cost = data.groupby('FlightDate')['Cost: Total $'].sum()
plt.figure(figsize=(10,5))
sns.lineplot(x = yearly_cost.values, y = yearly_cost.index, marker = 'o')
plt.title("Yearly Cost Incurred due to Bird Strikes")
plt.xlabel("Yearly Cost")
plt.ylabel("Year")
plt.grid()
plt.show()
```

# When do most bird strikes occur?

```python
flight_phase = data['When: Phase of flight'].value_counts()
plt.figure(figsize=(12,6))
plt.pie(
    flight_phase,
    labels = flight_phase.index,
    autopct='%1.1f%%',
    colors = sns.color_palette('Set1', len(flight_phase)),
)
plt.title('When do most bird strikes occur?')
plt.show()
```

# Altitude of aeroplanes at the time of strike

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['Feet above ground'], bins = 30)
plt.title('Altitude of aeroplanes at the time of strike')
plt.xlabel('Altitude(In Feet)')
plt.ylabel('Number of Bird Strikes')
plt.grid()
plt.show()
```

# Phase of flight at the time of the strike.

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='When: Phase of flight', data=data, order=data['When: Phase of flight'].value_counts().index)
plt.title('Distribution of Bird Strikes Across Different Phases of Flight')
plt.xlabel('Phase of Flight')
plt.ylabel('Number of Bird Strikes')
plt.show()
```

# Average Altitude of the aeroplanes in different phases at the time of strike


```python
plt.figure(figsize=(12, 6))
avg_altitude = data.groupby('When: Phase of flight')['Feet above ground'].mean()
sns.barplot(x = avg_altitude.index, y = avg_altitude.values, palette = "bone")
plt.title('Average Altitude of the aeroplanes in different phases at the time of strike')
plt.xlabel('Phase of Flight')
plt.ylabel('Altitude')
plt.xticks(rotation = 45)
plt.show()
```

# Effect of Bird Strikes & Impact on Flight


```python
plt.figure(figsize=(12, 6))
flight_impact = data['Effect: Impact to flight'].value_counts()
plt.pie(
    flight_impact,
    labels = flight_impact.index,
    autopct='%1.1f%%',
    colors = sns.color_palette('Dark2', len(flight_impact))
)
plt.legend(
    flight_impact.index,
    title = 'Effect of Flight',
    bbox_to_anchor = (1.4, 1),
    loc = 'upper right'
)
plt.show()
```

# Effect of Strike at Different Altitude

```python
plt.figure(figsize=(10, 5))
sns.countplot(x = 'Effect: Impact to flight', data=data)
plt.title('Effect of Strike at Different Altitude')
plt.xlabel('Impact of Flight')
plt.ylabel('Bird Strikes')
plt.xticks(rotation = 45)
plt.show()
```

# Were Pilots Informed? & Prior Warning and Effect of Strike Relation

```python
plt.figure(figsize=(12, 7))
sns.countplot(x='Effect: Impact to flight', hue='Pilot warned of birds or wildlife?', data=data)
plt.title('Effect of Bird Strikes by Prior Warning')
plt.xlabel('Impact to Flight')
plt.ylabel('Number of Bird Strikes')
plt.xticks(rotation=45)
plt.legend(title='Pilots Informed')
plt.show()
```

## Code in Modular Format :-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def my_data(file_path):
    data = pd.read_excel(file_path)
    return data
```

```python
def strikes_over_years(data):
    data['year'] = pd.to_datetime(data['FlightDate']).dt.year
    yearly_strikes = data['year'].value_counts().sort_index()
    plt.figure(figsize=(12, 6))
    sns.lineplot(
        x = yearly_strikes.index,
        y = yearly_strikes.values,
        marker = 'o',
        markersize=10,
        markerfacecolor='red'
    )
    plt.title('Yearly Bird Strikes')
    plt.xlabel('Year')
    plt.ylabel('Number of Bird Strikes')
    plt.show()
```

```python
def strikes_in_US(data):
    data['FlightDate'] = pd.to_datetime(data['FlightDate'])

    data['Year'] = data['FlightDate'].dt.year
    
    non_us_states = ['Quebec', 'British Columbia', 'Prince Edward Island', 'Ontario', 
                      'Newfoundland and Labrador', 'Alberta', 'Saskatchewan']
    
    data_us = data[~data['Origin State'].isin(non_us_states)]  
    
    
    plt.figure(figsize=(16, 8))
    sns.countplot(x='Year', hue='Origin State', data=data_us, palette='Paired')
    plt.title('Yearly Analysis & Bird Strikes in the US')
    plt.xlabel('Year')
    plt.ylabel('Number of Bird Strikes')
    plt.legend(title='Origin State', bbox_to_anchor=(1.03, 1.4))
    plt.show()
```

```python
def top_airline_strikes(data):
    top_10_airlines = data["Aircraft: Airline/Operator"].value_counts().head(10)
    plt.figure(figsize=(10,5))
    sns.barplot(x = top_10_airlines.index, y = top_10_airlines.values, palette='Blues_d')
    plt.title("Top 10 US Airlines in terms of having encountered bird strikes")
    plt.xlabel("Airlines")
    plt.ylabel("Number of Bird Strikes")
    plt.xticks(rotation=90)
    plt.show()
```

```python
def airports_top50(data):
    top_50 = data["Airport: Name"].value_counts().nlargest(50)
    plt.figure(figsize=(14,17))
    sns.barplot(x = top_50.values, y = top_50.index, palette = "CMRmap")
    plt.title("Airports with most incidents of bird strikes – Top 50")
    plt.xlabel("Number of Bird Strikes")
    plt.ylabel("Airports")
    plt.show()
```

```python
def yearly_cost_incurred(data):
    yearly_cost = data.groupby('FlightDate')['Cost: Total $'].sum()
    plt.figure(figsize=(10,5))
    sns.lineplot(x = yearly_cost.values, y = yearly_cost.index, marker = 'o')
    plt.title("Yearly Cost Incurred due to Bird Strikes")
    plt.xlabel("Yearly Cost")
    plt.ylabel("Year")
    plt.grid()
    plt.show()
```

```python
def altitude_of_airplanes(data):
    plt.figure(figsize=(12, 6))
    sns.histplot(data['Feet above ground'], bins = 30)
    plt.title('Altitude of aeroplanes at the time of strike')
    plt.xlabel('Altitude(In Feet)')
    plt.ylabel('Number of Bird Strikes')
    plt.grid()
    plt.show()
```

```python
def phase_of_flight(data):
    plt.figure(figsize=(12, 6))
    sns.countplot(x='When: Phase of flight', data=data, order=data['When: Phase of flight'].value_counts().index)
    plt.title('Distribution of Bird Strikes Across Different Phases of Flight')
    plt.xlabel('Phase of Flight')
    plt.ylabel('Number of Bird Strikes')
    plt.show()
```

```python
def average_altitude_different_phases(data):
    plt.figure(figsize=(12, 6))
    avg_altitude = data.groupby('When: Phase of flight')['Feet above ground'].mean()
    sns.barplot(x = avg_altitude.index, y = avg_altitude.values, palette = "bone")
    plt.title('Average Altitude of the aeroplanes in different phases at the time of strike')
    plt.xlabel('Phase of Flight')
    plt.ylabel('Altitude')
    plt.xticks(rotation = 45)
    plt.show()
```

```python
def effect_and_impact_on_flight(data):
    plt.figure(figsize=(12, 6))
    flight_impact = data['Effect: Impact to flight'].value_counts()
    plt.pie(
        flight_impact,
        labels = flight_impact.index,
        autopct='%1.1f%%',
        colors = sns.color_palette('Dark2', len(flight_impact))
    )
    plt.legend(
        flight_impact.index,
        title = 'Effect of Flight',
        bbox_to_anchor = (1.4, 1),
        loc = 'upper right'
    )
    plt.show()
```
